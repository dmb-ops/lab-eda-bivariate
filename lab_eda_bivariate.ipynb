```python
import pandas as pd 
import numpy as np 

file_path = ("/Users/dmb/Desktop/IRONHACK/week5/day2/amz_uk_price_prediction_dataset.csv")
df = pd.read_csv(file_path)
df
```

```python
def best_seller_crosstab(df):
    ctab = pd.crosstab(df['category'], df['isBestSeller'], normalize='index')     #1. Crosstab and Best-Seller Proportion
    return ctab.sort_values(True, ascending=False)  # proportion of True
```

```python
from scipy.stats import chi2_contingency
import numpy as np

def chi_square_and_cramers_v(df):
    table = pd.crosstab(df['category'], df['isBestSeller'])
    chi2, p, dof, _ = chi2_contingency(table)
    n = table.sum().sum()
    cramers_v = np.sqrt(chi2 / (n * (min(table.shape) - 1)))
    return {'chi2': chi2, 'p_value': p, 'cramers_v': cramers_v}   #Chi-Square Test and Cramér's V
```

```python
import matplotlib.pyplot as plt        #3. Stacked Bar Chart

def plot_stacked_bar_best_seller(df):
    table = pd.crosstab(df['category'], df['isBestSeller'])
    table.div(table.sum(1), axis=0).plot(kind='bar', stacked=True, figsize=(12, 6))
    plt.title('Proportion of Best Sellers by Category')
    plt.ylabel('Proportion')
    plt.xlabel('Category')
    plt.tight_layout()
    plt.show()           
```

2: Prices and Ratings Across Categories

```python
def remove_price_outliers(df):                #Remove Outliers using IQR
    Q1 = df['price'].quantile(0.25)
    Q3 = df['price'].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df['price'] >= Q1 - 1.5 * IQR) & (df['price'] <= Q3 + 1.5 * IQR)]
```

```python
import seaborn as sns      #Violin Plot of Prices by Top 20 Categories

def violin_plot_prices(df):
    top_categories = df['category'].value_counts().head(20).index
    filtered = df[df['category'].isin(top_categories)]
    plt.figure(figsize=(15, 6))
    sns.violinplot(x='category', y='price', data=filtered)
    plt.xticks(rotation=90)
    plt.title('Price Distribution by Category')
    plt.tight_layout()
    plt.show()
```

```python
def avg_price_bar_chart(df):      #Bar Chart of Avg Price by Top 10 Categories
    avg_price = df.groupby('category')['price'].mean()
    top_10 = df['category'].value_counts().head(10).index
    avg_price[top_10].sort_values(ascending=False).plot(kind='bar', figsize=(10, 5))
    plt.title('Average Price by Top 10 Categories')
    plt.ylabel('Average Price')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
def boxplot_ratings(df):    #Box Plot of Ratings by Top 10 Categories
    top_categories = df['category'].value_counts().head(10).index
    filtered = df[df['category'].isin(top_categories)]
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='category', y='stars', data=filtered)
    plt.xticks(rotation=90)
    plt.title('Rating Distribution by Category')
    plt.tight_layout()
    plt.show()
```

3: Relationship Between Price and Rating

```python
def price_rating_correlation(df):
    return df[['price', 'stars']].corr().loc['price', 'stars']
```

```python
def scatter_price_rating(df):  #Scatter Plot of Rating vs Price
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='stars', y='price', data=df, alpha=0.3)
    plt.title('Price vs Rating')
    plt.tight_layout()
    plt.show()
```

```python
def correlation_heatmap(df):   #Correlation heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.tight_layout()
    plt.show()
```

```python
import scipy.stats as stats                   #QQ Plot for Normality Check

def qq_plot_price(df):
    stats.probplot(df['price'], dist='norm', plot=plt)
    plt.title('QQ Plot of Product Prices')
    plt.tight_layout()
    plt.show()
```
